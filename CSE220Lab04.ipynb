```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:

  def __init__(self, a):
       self.head=Node(None,None,None)
       node=None
       self.arr=a
       if type(a)==list:
        for i in a:
                new_node=Node(i,None,None)
                if node==None:
                    self.head.next=self.head.prev=new_node
                    new_node.next=new_node.prev=self.head
                    node=new_node
                else:
                    node=self.head.prev
                    self.head.prev=node.next=new_node
                    new_node.prev=node
                    new_node.next=self.head

  def countNode(self):

    temp=self.head
    count=0
    for i in range(len(self.arr)):
      temp=temp.next
      count+=1
    return count

  def forwardprint(self):

    temp=self.head.next
    for i in range(len(self.arr)):
        if temp.next==self.head:
          print(temp.element)
        else:
          print(temp.element,end=',')
        temp=temp.next

  def backwardprint(self):

    temp=self.head.prev
    for i in range(len(self.arr)):
      if temp.prev==self.head:
        print(temp.element)
      else:
        print(temp.element,end=",")
      temp=temp.prev

  def nodeAt(self, idx):

    if idx<0 or idx>=len(self.arr):
      return None
    else:
      temp=self.head.next
      for i in range(idx):
        temp=temp.next
      return temp

  def indexOf(self, elem):

      temp=self.head.next
      count=0
      while temp!=self.head:
        if temp.element==elem:
          return count
        else:
          temp=temp.next
          count+=1
      return -1


  def insert(self, elem,idx):
    pass


  def remove(self, idx):
    pass
```
